# Step 7: Select input features

````python
X = df[
    [
        "Avg_Rating",
        "Complaints",
        "Delivery_Time_Min",
        "Customer_Loyalty_Score",
        "Service_Quality_Score",
        "Complaint_Rate_Pct",
        "Repeat_Customer_Rate_Pct",
        "Revenue_Per_Customer",
        "Revenue_Growth_Pct",
        "Repeat_Customer_Growth_Pct"
    ]
]
````

# Step 8: Select target

````python
y = df["Churn_Risk"]
````

# Step 8: Encode Target

````python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

y = le.fit_transform(y)

print(
    dict(
        zip(
            le.classes_,
            le.transform(le.classes_)
        )
    )
)
````

# Step 9 : Split data into train and testing data 

````python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print("Training Rows:", X_train.shape[0])
print("Testing Rows :", X_test.shape[0])
````

# Step 10:Verify Distribution

````python
import pandas as pd

print(
    pd.Series(y_train).value_counts(normalize=True)*100
)
````

````python
print(
    pd.Series(y_test).value_counts(normalize=True)*100
)
````

# Step 11 : Now Train The Model

````python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(
    n_estimators=200,
    random_state=42
)
````

````python
model
````

````python
model.fit(
    X_train,
    y_train
)
````

# Step 12: Make Predictions

````python
y_pred = model.predict(
    X_test
)
````

# Step 13: Evaluate

````python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(
    y_test,
    y_pred
)

print(
    "Accuracy:",
    round(
        accuracy * 100,
        2
    ),
    "%"
)
````

````python
from sklearn.metrics import classification_report

print(
    classification_report(
        y_test,
        y_pred
    )
)
````

````python
from sklearn.metrics import confusion_matrix

print(
    confusion_matrix(
        y_test,
        y_pred
    )
)
````

# Step 15b: Predict churn risk 

````python
scenario_predictions = model.predict(scenario_data)

scenario_predictions_text = le.inverse_transform(scenario_predictions)

scenario_predictions_text
````

````python
def churn_risk_summary_table(df):

    churn_summary = (
        df.groupby("Branch_Name")
        .agg(
            Avg_Rating=("Avg_Rating", "mean"),
            Avg_Complaints=("Complaints", "mean"),
            Avg_Delivery_Time=("Delivery_Time_Min", "mean"),
            Avg_Repeat_Customer_Rate=("Repeat_Customer_Rate_Pct", "mean"),
            Latest_Churn_Risk=("Churn_Risk", "last")
        )
        .reset_index()
    )

    churn_summary["Avg_Rating"] = churn_summary["Avg_Rating"].round(2)
    churn_summary["Avg_Complaints"] = churn_summary["Avg_Complaints"].round(0).astype(int)
    churn_summary["Avg_Delivery_Time"] = churn_summary["Avg_Delivery_Time"].round(1)
    churn_summary["Avg_Repeat_Customer_Rate"] = churn_summary["Avg_Repeat_Customer_Rate"].round(2)

    def churn_reason(row):
        reasons = []

        if row["Avg_Rating"] < 4.0:
            reasons.append("Low customer rating")

        if row["Avg_Complaints"] > df["Complaints"].mean():
            reasons.append("High complaints")

        if row["Avg_Delivery_Time"] > df["Delivery_Time_Min"].mean():
            reasons.append("Slow delivery")

        if row["Avg_Repeat_Customer_Rate"] < df["Repeat_Customer_Rate_Pct"].mean():
            reasons.append("Low repeat customers")

        if len(reasons) == 0:
            return "Healthy customer retention"

        return ", ".join(reasons)

    def churn_action(risk):
        if risk == "High":
            return "Immediate recovery offers, complaint reduction, improve delivery speed"
        elif risk == "Medium":
            return "Monitor complaints weekly, push loyalty offers, improve service quality"
        else:
            return "Maintain service quality and use as benchmark branch"

    churn_summary["Main_Reason"] = churn_summary.apply(churn_reason, axis=1)
    churn_summary["Recommended_Action"] = churn_summary["Latest_Churn_Risk"].apply(churn_action)

    churn_summary = churn_summary[
        [
            "Branch_Name",
            "Latest_Churn_Risk",
            "Avg_Rating",
            "Avg_Complaints",
            "Avg_Delivery_Time",
            "Avg_Repeat_Customer_Rate",
            "Main_Reason",
            "Recommended_Action"
        ]
    ]

    return churn_summary

````
